```python
from flask import Flask, render_template
import sqlite3 
import csv 
import os
```

```python
con = sqlite3.connect('Videogames.db')
cursor = con.cursor()
```

```python


create_table = """CREATE TABLE VideoGameStocks (
                    Date DATE,
                    Company VARCHAR(255),
                    Ticker_Symbol VARCHAR(50),
                    Currency VARCHAR(10),
                    Open FLOAT,
                    High FLOAT,
                    Low FLOAT,
                    Close FLOAT,
                    Adj_Close FLOAT,
                    Volume BIGINT
                );
                """
cursor.execute(create_table)
```

```python
with open("Top10VideoGameStocks.csv") as vg:
    data = csv.reader(vg)
    next(data)
    for row in data:
        cursor.execute("INSERT INTO VideoGameStocks VALUES (?,?,?,?,?,?,?,?,?,?)", (row[0], row[1], row[2], row[3], float(row[4]), float(row[5]), float(row[6]), float(row[7]), float(row[8]), int(float(row[9]))))

```

```python
cursor.execute("SELECT * FROM VideoGameStocks").fetchall()
```
